```python
from random import randint
```

```python
# счетчик попыток
c = 0

# датчик правильного ответа
answer = False

# количество попыток
attempts = int(input('Введите количество попыток: '))

# загадываем число
number = randint(1, 100)
print('Я загадал число от 1 до 100. Попробуй угадать его!', '\n')

# проверка вводимых чисел
while c < attempts:
  num_user = int(input('Введи число! '))
  if num_user > number:
    print('Загаданное число меньше введенного!')
  if num_user < number:
    print('Загаданное число больше введенного!')  
  if num_user == number:  
    print('Ты угадал число! Поздравляю!')
    answer = True
    break

  c += 1
  print('Ты использовал', c, 'попыток из', attempts, '\n')

# попытки закончились
if answer == False:
  print('Попытки закончились :( Загаданное число:', number)    
```
